```python
import pandas as pd
import matplotlib.pyplot as plt

# 리스트에 데이터 삽입하기
X = [1, 2, 3, 4, 5]
Y_linear = [2, 4, 6, 8, 10] # 선형 데이터

# 리스트를 데이터프레임으로 변환하기
data = {'X': X, 'Y_선형': Y_linear}
df = pd.DataFrame(data)
df
```


[lhm 9-2]

```python
# 그래프 그리기
df.plot(x='X', y='Y_선형' )
plt.show()
```


[lhm 9-3]

```python
#피어슨 상관분석 수행하기(X와 Y선형)
coef_p= df.corr(method='pearson')
print("피어슨 상관계수 (X와 Y_linear):")
print(coef_p)
```


[lhm 9-4]

```python
# 스피어만 상관분석 수행하기 (X와 Y선형)
coef_s = df.corr(method='spearman')
print("\n스피어만 상관계수 (X와 Y_linear):")
print(coef_s)
```


[lhm 9-5]

```python
# 켄달 상관분석 수행하기 (X와 Y선형)
coef_k = df.corr(method='kendall')
print("\n켄달 상관계수 (X와 Y_linear):")
print(coef_k)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
# 리스트에 데이터 삽입하기
X = [1, 2, 3, 4]
Y = [10, 9, 8, 7]

# 리스트를 데이터프레임으로 변환하기
data = {'XA': X, 'YA': Y}
df = pd.DataFrame(data)
df

# 그래프 그리기
df.plot(x='XA', y='YA')
plt.show()

# 피어슨 상관분석 수행하기 (A)
Acoef_p = df.corr(method='pearson')
print("피어슨 상관계수(A형):")
print(Acoef_p)

# 스피어만 상관분석 수행하기 (A)
Acoef_s = df.corr(method='spearman')
print("\n스피어만 상관계수 (A):")
print(Acoef_s)

# 켄달 상관분석 수행하기 (A)
Acoef_k = df.corr(method='kendall')
print("\n켄달 상관계수 (A):")
print(Acoef_k)
```
