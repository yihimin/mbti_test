```python
import os  # 운영 체제 관련 기능을 위한 모듈
import pandas as pd  # 데이터 처리를 위한 Pandas 라이브러리
import numpy as np  # 수치 계산을 위한 Numpy 라이브러리
import matplotlib.pyplot as plt  # 데이터 시각화를 위한 Matplotlib 라이브러리
import warnings  # 경고 메시지 제어를 위한 모듈
from datetime import datetime, date, time  # 날짜와 시간 처리를 위한 모듈
import csv  # CSV 파일 처리를 위한 모듈

# 데이터를 데이터프레임 형태로 읽어오기
df = pd.read_csv('lhm_대전태평(2009).csv', encoding='UTF-8')  # CSV 파일을 UTF-8 인코딩으로 불러옴
df.head(3)  # 데이터프레임의 첫 3개 행을 출력하여 데이터 확인

# 인덱스를 리셋하여 순차적으로 재설정
df = df.reset_index(drop=True)  # 기존 인덱스를 제거하고 0부터 순차적인 인덱스를 설정

# 날짜와 시간을 합쳐 새로운 'Date' 열 생성 후, 이를 인덱스로 설정
df['Date'] = pd.to_datetime(df['날짜'], format='%Y%m%d') + pd.to_timedelta(df['시간'].astype(int), unit='h')
df.set_index(df['Date'], inplace=True)  # 'Date' 열을 인덱스로 설정
df.head(3)  # 인덱스가 설정된 데이터프레임의 첫 3개 행을 출력하여 확인
```

```python
# 필요한 열만 선택하고, 열 이름을 변경하여 새로운 데이터프레임 생성
df1 = df[['수온(℃)', '수위(el.m)', 'EC(㎲/㎝)']]  # '수온', '수위', 'EC' 열을 선택
df1.columns = ['temp', 'level', 'EC']  # 열 이름을 간단하게 변경
df1.head(3)  # 첫 3개 행을 출력하여 변경된 열 이름을 확인
```

```python
# 결측치 개수를 확인하고 데이터를 CSV 파일로 저장
df1.isnull().sum()  # 각 열의 결측치 개수를 확인
```

```python
# 결측치가 있는 데이터프레임을 다시 불러와 기본 통계 정보를 확인
df = pd.read_csv('lhm_대전지하수.csv', index_col='Date', parse_dates=True, encoding='UTF-8')  # 저장된 CSV 파일을 다시 불러옴
df.describe()  # 각 열의 기본 통계 정보(count, mean, std, min, max 등)를 출력하여 데이터 분포 확인
```

```python
# 이상치 탐지를 위해 'level' 열의 IQR 계산
q3_level = df['level'].quantile(q=0.75)  # 3사분위수(Q3) 계산
q1_level = df['level'].quantile(q=0.25)  # 1사분위수(Q1) 계산
iqr_level = q3_level - q1_level  # IQR(Interquartile Range) 계산
print(iqr_level)  # IQR 값을 출력하여 확인

# 이상치 범위 계산: 1.5 * IQR을 이용해 상한과 하한 설정
upper_level = q3_level + 1.5 * iqr_level  # 상한값 계산
lower_level = q1_level - 1.5 * iqr_level  # 하한값 계산
print(upper_level, '/', lower_level)  # 상한과 하한 값 출력
print((df['level'] > upper_level).sum())  # 상한을 초과하는 이상치 개수 출력
print((df['level'] < lower_level).sum())  # 하한 미만의 이상치 개수 출력
```

```python
# 'level' 열에서 이상치를 제거하고 데이터 분포 시각화
df_iqr_level = df[(df['level'] < upper_level) & (df['level'] > lower_level)]  # 상한과 하한 범위 내의 값들만 선택
df_iqr_level['level'].plot()  # 이상치 제거 후 'level' 열의 데이터 분포를 시각화
plt.show()  # 그래프 출력
```

```python
# 'EC' 열의 IQR 계산 및 이상치 탐지
q3_ec = df['EC'].quantile(q=0.75)  # 'EC' 열의 3사분위수(Q3) 계산
q1_ec = df['EC'].quantile(q=0.25)  # 'EC' 열의 1사분위수(Q1) 계산
iqr_ec = q3_ec - q1_ec  # 'EC' 열의 IQR 계산
print('IQR:', iqr_ec)  # IQR 값 출력

# 'EC'의 상한과 하한 계산 및 이상치 개수 확인
upper_ec = q3_ec + iqr_ec * 1.5  # 'EC' 상한값
lower_ec = q1_ec - iqr_ec * 1.5  # 'EC' 하한값
print(upper_ec, '/', lower_ec)  # 상한과 하한 출력
print((df['EC'] > upper_ec).sum())  # 상한을 초과하는 'EC' 이상치 개수
print((df['EC'] < lower_ec).sum())  # 하한 미만의 'EC' 이상치 개수
```

```python
# 이상치를 제거한 'EC' 열의 분포 시각화
df_iqr_ec = df[(df['EC'] < upper_ec) & (df['EC'] > lower_ec)]  # 'EC'에서 이상치를 제거한 데이터프레임 생성
df_iqr_ec['EC'].plot()  # 'EC' 열의 이상치 제거 후 분포 시각화
plt.show()  # 그래프 출력
```

```python
# 'level'과 'EC'에서 동시에 이상치를 제거한 데이터프레임 생성 및 통계 정보 출력
df_iqr = df[(df['level'] < upper_level) & (df['level'] > lower_level) & (df['EC'] < upper_ec) & (df['EC'] > lower_ec)]
df_iqr.describe()  # 이상치 제거 후 데이터프레임의 통계 정보 출력
```

```python
# 이상치 제거 후 데이터 분포를 히스토그램으로 시각화
df_iqr.hist(bins=50, figsize=(10, 6))  # 히스토그램 생성
plt.show()  # 히스토그램 출력
```

```python
# 이상치 제거된 데이터를 CSV 파일로 저장
df_iqr.to_csv('lhm_대전지하수_이상치정제.csv', encoding='UTF-8')  # 정제된 데이터를 새로운 CSV 파일로 저장

# 이상치가 제거된 데이터 재불러오기 및 상자 그림 시각화
df = pd.read_csv('lhm_대전지하수_이상치정제.csv')
df.head()  # 첫 5개 행 출력

df.boxplot()  # 데이터 분포를 상자 그림(Boxplot)으로 시각화
```

```python
# 정규화를 위한 사본 생성 및 확인
df_copy1 = df.copy()  # 데이터프레임 사본 생성
df_copy1.head(3)  # 사본의 첫 3개 행 출력
```

```python
# 'level' 열의 Min-Max 정규화 수행
df_copy1['level_minmax'] = (df_copy1['level'] - df_copy1['level'].min()) / (df_copy1['level'].max() - df_copy1['level'].min())
df_copy1.head(3)  # 정규화 결과 확인
```

```python
# 정규화된 'level' 열의 히스토그램 시각화
df_level = df_copy1[['level', 'level_minmax']]  # 'level'과 'level_minmax' 열 선택
df_level.hist(figsize=(8, 6))  # 히스토그램 생성
plt.show()  # 히스토그램 출력
```

```python
# 'temp' 열의 Min-Max 정규화 수행 및 결과 확인
df_copy1['temp_minmax'] = (df_copy1['temp'] - df_copy1['temp'].min()) / (df_copy1['temp'].max() - df_copy1['temp'].min())
df_copy1.head(3)  # 정규화 결과 확인
```

```python
# 정규화된 'temp' 열의 히스토그램 시각화
df_temp = df_copy1[['temp', 'temp_minmax']]  # 'temp'와 'temp_minmax' 열 선택
df_temp.hist(figsize=(8, 6))  # 히스토그램 생성
plt.show()  # 히스토그램 출력
```
